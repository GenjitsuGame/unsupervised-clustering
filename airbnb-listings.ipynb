```python
import pandas as pd
```

```python
df_listings = pd.read_csv('./airbnb-listings.csv', sep=';')
```

```python
df_listings.columns
```

```python
features = [
    'Property Type', 'Room Type', 'Accommodates', 'Bathrooms', 'Bedrooms',
       'Beds', 'Bed Type', 'Amenities'
]
```

```python
df = df_listings.loc[:, features]
df.loc[:, 'Amenities'].fillna('', inplace=True)
df.dropna(how='any', inplace=True)
df = df.loc[df['Property Type'] != 'Other']
```

```python
df['Property Type'].value_counts()
```

```python
df
```

```python
amenities = df.Amenities[~df.Amenities.str.contains('{')].str.split(',').apply(lambda x: pd.Series(x)).stack()
amenities
```

```python
dummies = pd.get_dummies(amenities)
dummies
```

```python
encoded_amenities= dummies.groupby(level=[0]).apply(lambda x: x.sum(axis=0))
encoded_amenities
```

```python
df = df.join(pd.get_dummies(df.loc[:, ['Room Type', 'Bed Type']]))
df = df.join(encoded_amenities.iloc[:, 2:], how='inner')
df
```

```python
df.drop(columns=['Room Type', 'Bed Type', 'Amenities', 'translation missing: en.hosting_amenity_49',
       'translation missing: en.hosting_amenity_50'], inplace=True)
df
```

```python
df.loc[:, 'Property Type Labels'] = df['Property Type'].astype('category').cat.codes
```

```python
df.to_csv('embedded_listings.csv', sep=';', index=False)
```
